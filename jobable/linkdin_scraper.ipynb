```python
import json
import requests

def save_page_to_json(page_num):
  response = requests.get(f"https://jobdataapi.com/api/jobs/?region_id=5&max_age=30&page={page_num}")
  page_json = response.json()

  with open(f"api_page_{page_num}.json", mode="w") as f:
    json.dump(page_json, f)

start_page = 1
num_pages = 10

for page_num in range(start_page, start_page+num_pages):
  save_page_to_json(page_num)
```
