```python
import requests
from bs4 import BeautifulSoup

import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
url = 'https://news.ycombinator.com/item?id=47219667'
response = requests.get(url)
html_content = response.text
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
type(soup)
```

```python
print(soup.prettify())
```

```python
tag = soup.find('tr', class_='athing comtr')
```

```python
location = tag.find('div', class_='commtext c00')
type(location)
```

```python
for string in location.strings:
    print(string)
```

```python
df = {"self_description": ""}

for string in location.strings:
    chunks = string.split(' ')
    values = " ".join(chunks[1:])
    print(chunks)
    if chunks[0] == "Location:":
        df['location'] = values
    elif chunks[0] == "Remote:":
        df['remote'] = values
    elif chunks[0] == "Willing:":
        df['relocate_ok'] = values
    elif chunks[0] == "Technologies:":
        df['technologies'] = values
    elif chunks[0] in ["Resume:", "Résumé/CV:"]:
        pass
    elif chunks[0] == "Website:":
        pass
    elif chunks[0] == "Github:":
        pass
    elif chunks[0] == "Email:":
        pass
    elif chunks[0].endswith(":"):
        pass
    else:
        df['self_description'] += string + '\n' + " "
```
