```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, AutoTokenizer

# Trained Model ko load karo
model_path = "./cover_letter_model_2"
model = T5ForConditionalGeneration.from_pretrained(model_path, local_files_only=True)
# tokenizer = T5Tokenizer.from_pretrained(model_path, local_files_only=True)
```

```python
tokenizer = T5Tokenizer.from_pretrained("t5-small")
```

```python
def generate_cover_letter(job_title, qualifications, skills, experience):
    input_text = f"Job Title: {job_title}\nQualifications: {qualifications}\nSkills: {skills}\nExperience: {experience}"
    
    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)
    
    output = model.generate(**inputs, max_length=512, do_sample=True, top_k=50, top_p=0.95)
    cover_letter = tokenizer.decode(output[0], skip_special_tokens=True)
    
    return cover_letter

# Example Test
print(generate_cover_letter(
    "Data Scientist",
    "B.Tech in Computer Science",
    "Python, Machine Learning, Deep Learning",
    "2 years as a Data Scientist at XYZ"
))
```
